```python
import datetime
import pandas as pd
import numpy as np
```

```python
todays_date = datetime.datetime.now().date()
delta_days  = 3650 
date_range  = pd.date_range(todays_date - datetime.timedelta(delta_days), periods=delta_days+1, freq='D')
df          = pd.DataFrame(index=date_range, columns=['x','y']).fillna(0)
```

```python
#_dict = dict.fromkeys(pd.to_datetime(date_range), 'special day')
_dict = dict.fromkeys(date_range, 'special day')
type(date_range[0])

_lst = [np.datetime64(dt) for dt in date_range]

_dict = {np.datetime64(dt), 'special day' for }
```

```python
def check_day (row):
    '''
    es:
    df['special_day'] = df.apply(check_day, axis=1)
    df['special_day'] = df.apply(lambda row: check_day(row), axis=1)
    '''
    #key = pd.to_datetime(row.name)
    key = row.name   # type(row.name) = <class 'numpy.datetime64'>
    
    if key in _dict:
        return _dict[key]
    else:
        return str(type(row.name))
    
#
#

def check_day_iter (row):
    '''
    es:
    df['special_day'] = df.apply(check_day, axis=1)
    df['special_day'] = df.apply(lambda row: check_day(row), axis=1)
    '''
    key = pd.to_datetime(row.Index)
    if key in _dict:
        return _dict[key]
    else:
        return '-'
    #print(type(row))
```

```python
%%timeit
df['special_day'] = df.apply(lambda row: check_day(row), axis=1)
```

```python
%%timeit
for row in df.itertuples():
    df['special_day'] = check_day_iter(row)
```
