## talib candlestick pattern recognitron

```python
import numpy as np
import pandas as pd
import talib
import pprint
```

```python
pdf = pd.read_csv('/home/starq/REP/DATA/FINANCE/Quotazioni/ASA.csv' \
                                 ,usecols=['Date','Open', 'High', 'Low', 'Close', 'Volume'] \
                                 ,parse_dates=['Date'],infer_datetime_format=True \
                                 ,index_col='Date')
```

```python
help(talib.CDL3OUTSIDE)
```

```python
pdf['CDL2CROWS'] = talib.CDL2CROWS(pdf['Open'].values, pdf['High'].values, pdf['Low'].values, pdf['Close'].values)
```

```python
%timeit pdf['CDL2CROWS'] = talib.CDL2CROWS(\
                            pdf['Open'].values, pdf['High'].values, pdf['Low'].values, pdf['Close'].values)
```

```python
%timeit pdf.query('CDL2CROWS != 0')
```

```python
%timeit pdf.loc[pdf['CDL2CROWS'] != 0]
```

### BEST QUERY PERFORMANCE (np.where)

```python
%timeit pdf.iloc[np.where(pdf['CDL2CROWS'].values !=0)]
```

```python
help(talib.CDLENGULFING)
```

```python
pdf['CDLENGULFING'] = talib.CDLENGULFING(\
                        pdf['Open'].values, pdf['High'].values, pdf['Low'].values, pdf['Close'].values)
```

```python
%timeit pdf['CDLENGULFING'] = talib.CDLENGULFING(\
                        pdf['Open'].values, pdf['High'].values, pdf['Low'].values, pdf['Close'].values)
```

```python
%timeit pdf.iloc[np.where(pdf['CDLENGULFING'].values==100)]
```
